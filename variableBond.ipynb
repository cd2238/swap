```python
# list of Variables bonds
pd.set_option('display.max_rows', 300)
indexlist = []
col=['maturityDate', 'quotationdate', 'quotationvalue', 'currency', 'group', 'collateralType','callable', 'hybrid', 'inflationIndexed', 'index', 'marginadd'] #, 'index'
pdswap = pd.DataFrame(columns=col)
j=0
for i in range(len(dd)):
    if ((dd[i]['assetType']== typeAsset) and (dd[i]['interestRate'].get('index') != None)):
        #print('tt',dd[i]['interestRate'].get('index'))
        pdnew=pd.DataFrame([[dd[i]['maturityDate'],
                             dd[i]['quotations'][0]['date'],
                             dd[i]['quotations'][0]['close'],
                             dd[i]['ccy'], 
                             dd[i]['assetGroup'],
                             dd[i].get('collateralType'), 
                             dd[i]['callable'], 
                             dd[i]['hybrid'], 
                             dd[i]['inflationIndexed'], 
                             dd[i]['interestRate']['index']['_id'], 
                             dd[i]['interestRate']['marginAdd'], 
                             #dd[i]['legs'][0]['interest']['paymentFrequency']['code'], 
                             #dd[i]['legs'][0]['interest']['dayCountMethod'], 
                             #dd[i]['legs'][0]['interest']['index']['_id'], 
                             #dd[i]['legs'][1]['side'], 
                             #dd[i]['legs'][1]['interest']['rateType'], 
                             #dd[i]['legs'][1]['interest']['paymentFrequency']['code'], 
                             #dd[i]['legs'][1]['interest']['dayCountMethod'],
                             #dd[i]['legs'][1]['interest'].get('rate')
                            ]], 
                             columns=col, index=[dd[i]['name']])
        indexlist.append(dd[i]['interestRate']['index']['_id'])
        pdswap = pd.concat([pdswap, pdnew], ignore_index=False)
display(pdswap)
```
